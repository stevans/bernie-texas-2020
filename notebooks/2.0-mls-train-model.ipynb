```python
import pandas as pd
```

```python
#Read in all features
all_features = pd.read_csv('../data/interim/all_features.csv')
```

```python
all_features
```

```python
all_features['delta_fraction_sanders'] = all_features['fraction_sanders_2020'] - \
        all_features['fraction_sanders_2016']
```

```python
columns_to_train = ['fraction_white_nh', 'fraction_hispanic', 'fraction_black', \
                    'fraction_other_race', 'fraction_w4yrdeg', 'per_capita_income', \
                    'fraction_evangelical', 'fraction_catholic', 'fraction_mormon', \
                    'fraction_other_religion', 'margin_of_victory_D2016', 'population_density']
```

```python
column_target = ['delta_fraction_sanders']
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
data = all_features
X = data.loc[:,columns_to_train]  #independent columns
y = data.loc[:,column_target]    #target column i.e price range
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```
